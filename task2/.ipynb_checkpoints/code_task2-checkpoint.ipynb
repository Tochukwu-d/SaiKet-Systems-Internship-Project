```python
from pathlib import Path
import pandas as pd
```

```python
# load dataset into the pandas dataframe

DATA_PATH = Path("..\\data\\Telco_Customer_Churn_Dataset  (3).csv")
df = pd.read_csv(DATA_PATH)

# Confirm that the dataset is loaded
print("Dataset loaded successfullly....")
```

```python
# Standardize column name:
# Convert all to lowercase 

df.columns = df.columns.str.lower()
df.columns
```

```python
# Checking for missing values per columns

missing_values = df.isnull().sum()

missing_values
```

```python
# Separate and Classify columns into Numerical and Categorical

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns   # identify numerical columns

categorical_cols = df.select_dtypes(include=['object']).columns   # identify categorical columns
```
